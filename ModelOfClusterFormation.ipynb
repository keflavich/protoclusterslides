## Setting up an experiment

Starless cores (e.g. B68) and starless clumps (IRDCs) are well-studied and common.  The former are thought to be precursors to single stars, while the latter are likely to be precursors to open clusters.

Are there starless high-mass protoclusters, i.e. starless precursors to objects like NGC 3603, Westerlund 1, and The Arches? (these are known as "Young Massive Clusters" or YMCs)

New data - blind surveys of the Galactic plane - make this question approachable.

### What are the properties of a proto-YMC?

(this section could do with some references, but [Portegies-Zwart 2010](http://adsabs.harvard.edu/abs/2010ARA%26A..48..431P) contains most of the important details)

1. They are rare.  Their population follows a power law with index 2, such that a $10^4 M_\odot$ cluster is 100 times less likely than a $10^3 M_\odot$ cluster.  This is known from YMC studies in other galaxies (M83, M31).
2. They contain a lot of mass, by definition.  They need *at least* as much mass in gas as they'll have in stars, but probably $3-30\times$ more than that.
3. They are compact.  This is a tricky point: YMCs are observed to be very compact, with $r\sim1-5$ pc.  It is not possible to form a massive cluster with $r\sim10$ pc and have it shrink to $r\sim5$ pc over the lifetime of the most massive stars, yet the massive stars are preferentially seen twoards the center.  Therefore, the clumps that formed the massive stars must also form compact.
4. From #2 and #3 above, they must have high escape velocities.  They can therefore have large linewidths and yet remain bound.

### What are the *observable* properties of a proto-YMC?

Assuming they go through a starless phase, they will have $>10^4 M_\odot$ within a radius $r<2.5$ pc.

Existing surveys, specifically the [BGPS](irsa.ipac.caltech.edu/data/BOLOCAM_GPS/), [ATLASGAL](atlasgal.mpifr-bonn.mpg.de), and [Hi-Gal](hi-gal.iaps.inaf.it/wiki/doku.php), are complete to such objects throughout the Galactic plane.

Where do star clusters form?
----------------------------

Two simple models:        

 * Uniform disk: 
 
$p(r) = 1$ if $r_1 < r < r_2$

$p(r) = 0$ otherwise


 * Exponential disk: 
        
$p(r) = r_1 e^{-r}$ for $r>r_1$

$p(r) = 0$ for $r<r_1$

```python
figure(figsize=(8,6))
r = np.linspace(0,20,1000)
r_in = 3
r0 = 8.
p1 = ((r>r_in) * (r<15))/2.
p2 = (exp(-r/r0)/exp(-1/r0)*(r>r_in))
plot(r,p1,linewidth=2)
plot(r,p2,linewidth=2)
ylabel("Relative Probability Density")
xlabel("Galactocentric Radius (kpc)")
lims = axis([0,20,0,1.1])
```

Our sample is all within $r<8.5$ kpc.

What is their mass function?
----------------------------

A Schecter function?  Or a power-law?


The intrinsic mass function:
$$p(m) = A m^{-\beta} e^{-m/m_0} $$
where $m_0$ is a "maximum mass" above which the number of clusters exponentially drops off

```python
import imf
```

```python
m = np.linspace(1000,5e5)
p1 = imf.schechter(m,m0=3e4)
p2 = m**-2
```

```python
figure(figsize=(8,6))
ax=gca()
ax.grid(False)
ax.tick_params(axis='both', which='major', labelsize=20)
result = loglog(m,p1,linewidth=2)
result = loglog(m,p2,linewidth=2)
axis([m.min(),m.max(),1e-12,1e-6])
xl = xlabel("Mass",fontsize=25)
yl = ylabel("N(M) dM",fontsize=25)
```

However, we only examined clusters in the range $m>10^4$

How do the massive clusters form?
---------------------------------




 * Either all their gas is in place at once
    - call this the "prestellar core" model

 * Or it isn't
    - Some stars form, then others are accreted?
    - Or, some stars form, and *gas* is accreted
      + implies multiple generations
      + or "prolonged" star formation

Start with the first model: We can constrain it.
    
 * First, clusters go through a "starless" phase
 * Then, they form stars!
 * Finally, the stars expel the gas

The starry, gas-rich phase lasts at least until the first supernova if $v_{esc} > C_{II}$ 
 [(Bressert, Ginsburg et al 2012)](http://arxiv.org/abs/1208.3472) or until the gas is consumed

The first SN is at $t>3$ Myr.  The gas consumption timescale may be shorted.

We measure an upper limit: 
    
 $N_{starless}(M>10^4) = 0$
 
 $N_{starry}(M>10^4) = 3$ (conservatively) or $18$ (if you allow for 100% star formation efficiency)

The [Urquhart et al 2013](http://arxiv.org/abs/1302.2538) ATLASGAL sample extends this to ~6 in the conservative case.

 * By selecting on masers, that sample may be biased
 * Contreras and Longmore working on the unbiased sample?

So we "measure" $N_{starless}/N_{starry}$ = 0.

Poisson statistics: 
    $$\sigma_N = \frac{\sqrt{N_{starry}}}{N_{starry}} \approx 0.6 $$
    
Or, a $1-\sigma$ limit $N_{starless} < 0.6$

Lifetime of the starless phase:
    
 $$\frac{N_{starless}}{N_{starry}} = \frac{\tau_{starless}}{\tau_{starry}}$$
    
 $$ \tau_{starless} < \tau_{starry} \frac{\sqrt{N_{starry}}}{N_{starry}}  = 3 Myr * \frac{0.6}{3} = 0.6 Myr$$

Better statistics aren't warranted yet.

What have we learned about the formation of massive clusters?
-------------------------------------------------------------

The proto-massive-cluster "clumps" of gas form massive stars rapidly *if* they exist at all.

It is possible that a single large clump never exists as a precursor to a high-mass cluster.

In this case, they must be assembled from smaller sub-clumps or larger, more extended clouds.

These are sometimes referred to as the "cloud collapse" model and "assembly line" models.

"The Brick", the ~$10^5 M_\odot$ cloud in the CMZ, may form a cluster by an entirely different mechanism.

There is clearly environmental dependence for massive cluster formation.

Distances
---------
Velocity is enough to give an upper limit on distance.

```python
import kdist
lon = 10
vtan = kdist.kdist(lon,0,0,returnvtan=True)
v = np.linspace(-30,vtan,1000)
Dnear = kdist.kdist(lon*np.ones(1000),0*np.ones(1000),v,near=True,silent=True)
Dfar = kdist.kdist(lon*np.ones(1000),0*np.ones(1000),v,near=False,silent=True)
```

```python
figure(figsize=(10,8))
ax = gca()
plot(v,Dnear)
plot(v,Dfar)
ax.tick_params(axis='both', which='major', labelsize=20)
xlabel("$V_{LSR}$ (km/s)",fontsize=25)
ylabel("Distance (pc)",fontsize=25)
lims = axis([v.min(),v.max(),0,20e3],)
savefig("distance_vs_velocity_lon10.pdf",bbox_inches='tight')
```

For a given velocity, we have an upper limit on the distance.

We also have measurements of the velocities of all BGPS sources.

```python
getfile('http://eldora.as.arizona.edu/~yshirley/Arizona/BGPS/BGPS_vlsr.cat')
getfile('http://irsa.ipac.caltech.edu/data/BOLOCAM_GPS/tables/bolocam_gps_v1_0_1.tbl')
```

```python
import astropy.table
import astropy.coordinates
from astropy import units as u
T = astropy.table.Table.read('BGPS_vlsr.cat',format='ascii')
v = T['col7']
fk5 = astropy.coordinates.ICRS(T['col2']*u.deg, T['col3']*u.deg)
lon,lat = fk5.galactic.l.to(u.deg).value, fk5.galactic.b.to(u.deg).value
mask = (v<999)
v = v[mask]
lon = lon[mask]
lat = lat[mask]
dmax = kdist.kdist(lon,lat,v,near=False,silent=True)
bgps = astropy.table.Table.read('bolocam_gps_v1_0_1.tbl',format='ascii.ipac')
f40 = np.array([bgps['flux_40'][bgps['cnum'] == x] for x in T['col1'][mask]]).squeeze()
# Use standard conversion factors assuming T=20K, 1.5x beam correction & 1.5x flux correction, kappa = 0.0114 cm^2/g
mmax = 14.01 * f40 * 2.25 * (dmax/1e3)**2
```

Using these maximum velocities, we can acquire a mass-limited sample, as long as we can later disambiguate which of the potentially massive sources are genuinely massive (and distant). 

```python
figure(figsize=(16,8))
ax1 = subplot(1,3,1)
h,l,p = hist(v,bins=np.linspace(v.min(),v.max(),50))
ax2 = subplot(1,3,2)
h,l,p = hist(dmax/1e3,bins=np.linspace(0,20,50))
ax3 = subplot(1,3,3)
h,l,p = hist(mmax,bins=np.logspace(-0.25,5.5,50))
ax3.set_xscale('log')
for ax in ax1,ax2,ax3:
    ax.tick_params(axis='both', which='major', labelsize=20)
ax1.set_xlabel("$V_{LSR}$ (km/s)",fontsize=25)
ax2.set_xlabel("Max Distance (kpc)",fontsize=25)
ax3.set_xlabel("Max Mass ($M_{\odot}$)",fontsize=25)
savefig("velocity_maxdist_maxmass.pdf",bbox_inches='tight')
```

What is that excess at negative velocities?  Is it anything interesting?

```python
figure(figsize=(16,8))
ax1 = subplot(1,3,1)
h,l,p = hist(v[lon<90],bins=np.linspace(v.min(),v.max(),50))
h,l,p = hist(v[lon>90],bins=np.linspace(v.min(),v.max(),50))
ax2 = subplot(1,3,2)
h,l,p = hist(dmax[lon<90]/1e3,bins=np.linspace(0,20,50))
h,l,p = hist(dmax[lon>90]/1e3,bins=np.linspace(0,20,50))
ax3 = subplot(1,3,3)
h,l,p = hist(mmax[lon<90],bins=np.logspace(-0.25,5.5,50))
h,l,p = hist(mmax[lon>90],bins=np.logspace(-0.25,5.5,50))
ax3.set_xscale('log')
for ax in ax1,ax2,ax3:
    ax.tick_params(axis='both', which='major', labelsize=20)
ax1.set_xlabel("$V_{LSR}$ (km/s)",fontsize=25)
ax2.set_xlabel("Max Distance (kpc)",fontsize=25)
ax3.set_xlabel("Max Mass ($M_{\odot}$)",fontsize=25)
savefig("velocity_maxdist_maxmass_innerouter.pdf",bbox_inches='tight')
```

Those negative velocities mostly come from low-mass clumps in the outer galaxy; they're nothing special.
